```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option("max_rows", 20)
pd.set_option('display.float_format', lambda x: '%.5f' % x)

%matplotlib inline
import matplotlib.pyplot as mpl
mpl.rcParams['font.size'] = 30
import matplotlib.pyplot as plt
#plt.style.use(['ggplot','dark_background'])
plt.style.use('seaborn')
import seaborn as sns

from math import *
from math import sqrt, factorial
from itertools import product, combinations, permutations

import scipy as sp
import scipy.stats as stats

from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

**Load, prepare & transform the the Data for regression**

```python
care = pd.read_csv('../data/medicare_clean.csv',
                   converters={'ProviderZipCode':str,'Provider Id':str})

#log transfrom possible Y variable for regression
care['LogTotalPayments'] = np.log(care['AverageTotalPayments'])\
    .map(lambda x: np.nan if np.isneginf(x) else x) 

# log transfrom possibe Y variable for regression 
care['LogUncoveredPayments'] = np.log(\
    care['AverageTotalPayments'] - care['AverageMedicarePayments'])\
    .map(lambda x: np.nan if np.isneginf(x) else x) 

care.drop('Unnamed: 0', inplace=True, axis=1)

#clean DRG definition names
care.DRGDefinition = care.DRGDefinition.map(lambda x: x[6:])

care.dropna(inplace=True)

care.info()
```

```python
census = pd.read_csv('../data/census_clean.csv', converters={'Zipcode': str})
income = pd.read_csv('../data/MedianIncome.csv', converters={'ZipCode': str})

income.Zipcode = income.Zipcode.map(lambda x: str(x[6:]))

census = census.merge(income, left_on='Zipcode', right_on='Zipcode', how='left')

census.columns = census.columns.str.replace('_', '')

census.MedianIncome = census.MedianIncome.map(
    lambda x: np.nan if isinstance(x, str) else x).map(float)

census.drop('Unnamed: 0', inplace=True, axis=1)

census.info()
```

**Join datasets & create log transformed X variable dataset** 

```python
# log tranfrom all x variables
log_census = census.copy()

for col in log_census.columns[1:]:

    log_census[col] = log_census[col].apply(np.log)\
    .apply(lambda x: np.nan if ~np.isfinite(x) else x)

# covert negative infinity and drop values
log_census.dropna(inplace=True)
census.dropna(inplace=True)

census.info(), log_census.info()
```

## Left Join on Medicare Data

```python
df = care.merge(census, left_on='ProviderZipCode', right_on='Zipcode', how='left')
logdf = care.merge(log_census, left_on='ProviderZipCode', right_on='Zipcode', how='left')

# drop 9999999.99 Values
df = df.applymap(lambda x: np.nan if x == 9999999.99 else x)
logdf = logdf.applymap(lambda x: np.nan if x == 9999999.99 else x)

df.info()
```

## filter data set to reduce bias

```python
clist = care.groupby('DRGDefinition').sum()\
            .sort_values('TotalDischarges',ascending=False)\
            .head(25).index
```

**Subset of data to with highest frequencies**

```python
sdf = df[df.DRGDefinition.isin(clist)].copy()
logsdf = logdf[logdf.DRGDefinition.isin(clist)].copy()
```

**Further clean DRG definitions for graphing**

```python
def procedure_clean(string):
    definition = ' '
    string = string.replace('&', '').replace('W', '')
    
    for i in range(string.count(' ')):
        definition += string.split(' ')[i] + ' '
        
        if i >= 2:
            break
    if definition.isspace():
        return string.strip()
    else:
        return definition.strip()
```

```python
sdf.DRGDefinition = sdf.DRGDefinition.map(procedure_clean)
logsdf.DRGDefinition = logsdf.DRGDefinition.map(procedure_clean)
```

### Visualize Subset

```python
bar1 = sdf.groupby('DRGDefinition').mean()\
          .sort_values('AverageTotalPayments', ascending=False)\
          .head(10).reset_index()

fig = plt.figure(figsize=(14,7))

#sns.set(font_scale=1.6)

plt.rc('xtick', labelsize=18)
plt.rc('ytick', labelsize=18)

plt.barh(bar1.DRGDefinition, bar1.AverageTotalPayments, xerr=500)

plt.title('Most Expensive Common Procedures', fontsize=22)

plt.savefig('../img/Most Expensive Common Procedures.png', bbox_inches = "tight")
```

**Further reduce based on frequency**

```python
elist = ['HEART FAILURE',
         'PULMONARY EDEMA',
         'RENAL FAILURE',
         'SIMPLE PNEUMONIA',
         'PSYCHOSES']

sdf = sdf[sdf.DRGDefinition.isin(elist)].copy()
logsdf = logsdf[logsdf.DRGDefinition.isin(elist)].copy()
```

```python
logsdf.dropna(inplace=True)
sdf.dropna(inplace=True)

sdf.drop(['AverageCoveredCharges','AverageMedicarePayments'],
         axis=1, inplace=True)
logsdf.drop(['AverageCoveredCharges','AverageMedicarePayments'],
            axis=1, inplace=True)
```

**Add dummy variables for each categorical x variable**

```python
logsdf = pd.get_dummies(data=logsdf,columns=['DRGDefinition'],drop_first=True)
sdf = pd.get_dummies(data=sdf,columns=['DRGDefinition'],drop_first=True)
```

```python
logsdf.corr()
```

```python
X = sm.add_constant(sdf.iloc[:,[13,16,19,20,21,22,23]])
y = sdf.AverageTotalPayments

sdf_model = sm.OLS(y,X).fit()
sdf_model.summary2()
```

```python
sdf_model.pvalues
```

```python
fig, ax = plt.subplots(figsize=(12,8))
predictions = sdf_model.predict(X)
ax.scatter(predictions,
           sdf['AverageTotalPayments'] - predictions,
           alpha=0.2)
ax.axhline(0, color='r', ls='--')

ax.set_xlabel('Predicted Average Total Payments')
ax.set_ylabel('Residuals');
```

```python
fig, ax = plt.subplots(figsize=((10, 10)))
predictions = sdf_model.predict(X)
stats.probplot(y - predictions, plot=ax);
ax.set_title("QQ Plot vs Normal dist for Average Total Payments Residuals")
```

```python
logX = sm.add_constant(logsdf.iloc[:,[13,16,19,20,21,22,23]])
logy = logsdf.LogTotalPayments

logsdf_model = sm.OLS(logy,logX).fit()
logsdf_model.summary2()
```

```python
logsdf_model.pvalues
```

```python
fig, ax = plt.subplots(figsize=(12,8))
predictions = logsdf_model.predict(logX)
ax.scatter(predictions,
           logy - predictions,
           alpha=0.2)
ax.axhline(0, color='r', ls='--')
ax.set_xlabel('Predicted Log Total Payments',fontsize=20)
ax.set_ylabel('Residuals',fontsize=30);
ax.set_title('Diagnostic Model Residual Plot',fontsize=20)

fig.savefig('../img/Residual plot', bbox_inches = "tight")
```
